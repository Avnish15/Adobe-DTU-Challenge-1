## Text Processing

```python
import re
import nltk
import spacy
import string
```

```python
# Calculating the length of the content
df['content_length'] = df['content'].apply(lambda text : len(text))
df.head()
```

```python
# Lower Casing : Convert the input text into same casing format
df['content'] = df['content'].str.lower()
df.head()
```

```python
# Punctuations Removal : a text standardization process
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

df["content"] = df['content'].apply(lambda text: remove_punctuation(text))
df.head()
```

```python
# Stopword Removal : words in a language like 'the', 'a' and so on  can be removed from the
# text most of the times, as they don't provide valuable information for downstream analysis.

nltk.download('stopwords')

from nltk.corpus import stopwords

STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

df["content"] = df["content"].apply(lambda text: remove_stopwords(text))
df.head()
```

```python
# Frequent Words
from collections import Counter
cnt = Counter()
for text in df['content'].values:
    for word in text.split():
        cnt[word] += 1

cnt.most_common(10)
```

```python
# Rare Words
n_rare_words = 10
RAREWORDS = set([w for (w, wc) in cnt.most_common()[:-n_rare_words-1:-1]])
RAREWORDS
```

```python
# Stemming : If there are two words in the corpus walks and walking, then stemming will stem the suffix to make them walk.
from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

df["content"] = df["content"].apply(lambda text: stem_words(text))
df.head()
```

```python
nltk.download('all')
```

```python
def remove_emoji(string):
  emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
  return emoji_pattern.sub(r'', string)

df['content'] = df['content'].apply(lambda text : remove_emoji(text))
df.head()
```

```python
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

df['content'] = df['content'].apply(lambda text : remove_urls(text))
df.head()
```

```python
def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)

df['content'] = df['content'].apply(lambda text : remove_html(text))
df.head()
```

### Bivariate Analysis

```python
from scipy import stats

# Pearson Test
output_df = pd.DataFrame(columns=['Correlation coefficient', 'p-value'])

for col in df :
  if not col == 'likes' :
    if df[col].isnull().sum() == 0 :
      if pd.api.types.is_numeric_dtype(df[col]) :
        r, p = stats.pearsonr(df['likes'], df[col])
        output_df.loc[col] = [round(r, 5), round(p, 6)]


output_df.reindex(output_df['Correlation coefficient'].abs().sort_values(ascending = False).index)
```

##### 1. It suggests that there is almost no linear relationship between the content length and likes. And it is statistically significant since p is very small.

```python
# Extracting features from timestamp object
df['day'] = df['date'].dt.day
df['week'] = df['date'].dt.isocalendar().week
df['hour'] = df['date'].dt.hour
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
df.head()
```

```python
output_df = pd.DataFrame(columns=['F-statistic', 'p-value'])

Categorical_columns = ['media_type', 'hour', 'day', 'week', 'month', 'year', 'username', 'inferred company']

for col in Categorical_columns :

  groups = df[col].unique()
  df_grouped = df.groupby(col)
  group_labels = []
  for g in groups :
    g_list = df_grouped.get_group(g)
    group_labels.append(g_list['likes'])
  f, p = stats.f_oneway(*group_labels)
  output_df.loc[col] = [round(f, 5), round(p, 6)]


output_df.reindex(output_df['F-statistic'].abs().sort_values(ascending = False).index)
```

###Word-to-Vec

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import gensim.downloader as api
```

```python
stop_words = set(stopwords.words('english'))
df['tokenized_text'] = df['content'].apply(lambda x: [w.lower() for w in word_tokenize(x) if w.isalpha() and w.lower() not in stop_words])
```

```python
model = Word2Vec(df['tokenized_text'], vector_size=100, window=5, min_count=1, workers=2)
```

```python
def text_to_vector(text):
    tokens = [w.lower() for w in word_tokenize(text) if w.isalpha() and w.lower() not in stop_words]
    vector = sum([model.wv[word] for word in tokens if word in model.wv])
    return vector.tolist() if vector.any() else [0] * model.vector_size
```

```python
df['text_vectors'] = df['content'].apply(text_to_vector)
```

```python
def extract_urls(link):

  if link.find('Photo') != -1 :
    start_index = link.find("previewUrl='") + len("previewUrl='")
    end_index = link.find("'", start_index)
    media_url = link[start_index:end_index]

  if link.find('Video') != -1 :
    start_index = link.find("thumbnailUrl='") + len("thumbnailUrl='")
    end_index = link.find("'", start_index)
    media_url = link[start_index:end_index]

  if link.find('Gif') != -1 :
    start_index = link.find("thumbnailUrl='") + len("thumbnailUrl='")
    end_index = link.find("'", start_index)
    media_url = link[start_index:end_index]


  return media_url
```
